**SP12, #5 (Spin)**

*Problem:* Consider a system of two spins that are coupled via a spin-spin interaction $W=a\mathbf{S}_1\cdot\mathbf{S}_2$.  Here $a$ is a measure of the coupling strength.  How many (spin) states does the two-particle system have, and what are their energy levels?  Show how you calculate this.

*Solution:* There are four possible states:

**Triplet states** ($S=1$):
\begin{eqnarray}
\lvert1,1\rangle &= \lvert\uparrow\uparrow\rangle \newline
\lvert1,0\rangle &= \frac{1}{\sqrt{2}} \left(\lvert\uparrow\downarrow\rangle + \lvert\downarrow\uparrow\rangle\right) \newline
\lvert1,-1\rangle &= \lvert\downarrow\downarrow\rangle
\end{eqnarray}

**Singlet state** ($S=0$):
$$\lvert0,0\rangle = \frac{1}{\sqrt{2}}\left(\lvert\uparrow\downarrow\rangle - \lvert\downarrow\uparrow\rangle\right) $$

The interaction Hamiltonian is $a \mathbf{S}_1 \cdot \mathbf{S}_2 = a\left(S_x^{(1)}S_x^{(2)} + S_y^{(1)}S_y^{(2)} + S_z^{(1)}S_z^{(2)}\right)$; so from the Schrodinger equation:
\begin{align*}
W\lvert\uparrow\uparrow\rangle &= a\left(S_x^{(1)}S_x^{(2)} + \ldots\right) \lvert\uparrow\uparrow\rangle \newline
&= a\frac{\hbar^2}{4} \left(\lvert\downarrow\downarrow\rangle + (i)(i)\lvert\downarrow\downarrow\rangle + \lvert\uparrow\uparrow\rangle\right) \newline
&= a\frac{\hbar^2}{4} \lvert\uparrow\uparrow\rangle
\end{align*}
and it can be shown for the other triplet states that the result of $W\lvert\Psi\rangle$ is the same, so the $S=1$ states all have energy $E_{trip}=a\hbar^2/4$.

By similarly applying $W$ to the singlet state, one finds that the energy is $-3a\hbar^2/4$.

**SP12, #6 (Reflection/transmission)**

*Problem:* A one-dimensional non-relativistic particle interacts with the potential $V=\lambda\frac{\hbar^2}{2m}\delta(x)$.

1. Calculate the reflection and transmission coefficients (probabilities) as a function of the incident particle wavenumber $k$.
2. Calculate the scattering and bound states for $\lambda < 0$.  Show that there is a single bound state, and that it is orthogonal to the scattering states.

*Solution*:
The Schrodinger equation is solved in the regions $x\neq0$ by the general plane-wave states
\begin{align*}
\psi(x) = \begin{cases}
A e^{ikx} + B e^{-ikx} & \text{if } \; x < 0 \newline
C e^{ikx} + D e^{-ikx} & \text{if } \; x > 0
\end{cases}
\end{align*}

One boundary condition is that the wavefunction be continuous at $x=0$ (as it should be everywhere); this enforces $\boxed{A + B = C + D}$.  The second boundary condition we use is the discontinuity in $\psi^{\prime}(x=0)$, which is found by integration of the S.E.  Namely, $\partial_x\psi(x=0^+) - \partial_x\psi(x=0^-) = -\lambda\psi(x=0)$.  Substituting in our wavefunction, this gives the second boundary condition, $\boxed{-\lambda(A+B) = ik(C-D-A+B)}$.

Let's find the bound state, which has $E<0$.  In the region $x<0$, the S.E. reads:
$$ \partial_x^2 \psi(x) = \kappa^2 \psi(x); \quad \kappa \equiv \frac{\sqrt{-2mE}}{\hbar} $$
The general solution to this equation is $\psi(x) = Ae^{\kappa x} + Be^{-\kappa x}$; we have to set $B=0$ so that $\psi$ doesn't blow up at $x=-\infty$.  A similar argument holds for the region $x>0$.  Using the first boundary condition,
$$
\psi(x) = \begin{cases}
Ae^{\kappa x}, & x<0 \newline
Ae^{-\kappa x}, & x>0
\end{cases}
$$
The second boundary condition (using the above form for the wavefunctions) gives $\lambda = 2\kappa$.  Finally, to eliminate $B$, normalize the wavefunction; this gives a bound state
$$
\psi(x) = \begin{cases}
\sqrt{-\lambda/2}e^{-\lambda x/2}, & x<0 \newline
\sqrt{-\lambda/2}e^{\lambda x/2}, & x>0
\end{cases}
$$
(note the choice of signs since $\lambda<0$).

The scattering states (for which $E>0$) satisfy $\partial_x^2 \psi(x) = -k^2 \psi(x)$, where $k \equiv \sqrt{2mE}/\hbar$.  Unforunately, the plane wave states are not normalizable, so there will always be one constant we can't eliminate.  However, we can at least get rid of $D$ if we imagine sending in particles from $x=-\infty$, and none from the right.  The best we can do, then, is write
$$
\psi(x) = \begin{cases}
A\left(e^{ikx} + \frac{i\lambda/2k}{1-i\lambda/2k}e^{-ikx}\right), & \text{if }\;x<0 \newline
A\left(1 + \frac{i\lambda/2k}{1-i\lambda/2k}\right)e^{ikx}, &\text{if }\;x>0
\end{cases}
$$

After a lot of algebra, we can solve for the reflection and transmission coefficients:
\begin{align*}
R(k) &\equiv \frac{\lvert B \rvert^2}{\lvert A \rvert^2} = \frac{(\lambda/k)^2}{1+(\lambda/k)^2} \newline
T(k) &\equiv \frac{\lvert C \rvert^2}{\lvert A \rvert^2} = \frac{1}{1+(\lambda/k)^2}
\end{align*}

**SP11, #5 (Entanglement)**

*Problem:* Consider two entangled spin $1/2$ particles separated by a large distance.  The four Bell states are given as
\begin{align*}
\lvert\Phi_{\pm}^{(AB)}\rangle &= \frac{1}{\sqrt{2}} \left[\lvert z+,z+\rangle \pm \lvert z-,z-\rangle\right] \newline
\lvert\Psi_{\pm}^{(AB)}\rangle &= \frac{1}{\sqrt{2}} \left[\lvert z+,z-\rangle \pm \lvert z-,z+\rangle\right] \newline
\end{align*}
where $\lvert z\pm\rangle$ are the two states along or against the $z$ axis; the first slot refers to particle $A$ which is nearby, the second to particle $B$ which is far away.  Consider applying a magnetic field only at location $A$, either in the $\hat{x}$ or $\hat{z}$ directions, for a time $t$.

a. Write down the Hamiltonians, $H_z$ and $H_x$, for the two operations above, and also write down the time evolution operators $U_z(t,0)$ and $U_x(t,0)$.

b. Find a time duration $t_A$ which results in $U_z\lvert z+\rangle = -i\lvert z+\rangle$.

c. Using time $t_A$, find $U_z\lvert z-\rangle$ and $U_x\lvert z\pm\rangle$.  (Hint: you may need to change from the $z$ basis to the $x$ basis).

d. Using the two operations above, show that any one of the Bell states can be transformed into any other, when operating only at location $A$.

*Solution:*

a. The Hamiltonian for a spin in an external field is $H=-\vec{\mu}\cdot\vec{B}$.  Since only particle $A$ feels the external field, $H_x = -\omega S_x$ (with $\vec{B}$ applied in the $x$ direction) and $H_z = -\omega S_z$, where $\omega = qB_{(x,z)}/mc$ is the gyromagnetic frequency.

The time evolution operator is $U(t,0)=e^{-iHt/\hbar}$; therefore $U_x(t,0) = e^{i\omega S_x t/\hbar}$ and $U_z(t,0) = e^{i\omega S_z t/\hbar}$.

b. $U_z(t,0)\lvert z+\rangle = e^{i\omega t/2}\lvert z+\rangle$; we want $e^{i\omega t_A/2}=-1$, so $t_A = \frac{(2n+1)2\pi}{\omega}$ for arbitrary $n\in\mathbb{Z}$.

c. $U_z(t_A,0)\lvert z-\rangle = e^{-3i\pi/2}\lvert z-\rangle = i\lvert z-\rangle$.

Rewrite $\lvert z\pm\rangle = \frac{1}{\sqrt{2}} \left[\lvert x+\rangle \pm \lvert x-\rangle\right]$ to find that $U_x(t_A,0)\lvert z\pm\rangle = i\lvert z\mp\rangle$.

d. For example, $U_z(t_A,0)\lvert\Phi_+^{(AB)}\rangle = -i\lvert\Phi_-^{(AB)}\rangle$, and so on.

**SP11, #6 (Perturbation theory)**

*Problem:*

a. The ground state wave function of the one dimensional harmonic oscillator ($H_{harm}=\frac{p^2}{2m}+\frac{1}{2}m\omega^2 x^2$) has the Gaussian form $\Psi(x)=Ce^{-k^2 x^2}$.  Using the Schrodinger equation, find $k$, and also find the ground state energy $E_0$.  Show all steps.

b. Use first order perturbation theory to find the first order correction $\Delta E$ to the ground state energy given by a perturbation to the potential, $H_{pert}=\lambda \lvert x \rvert$, with $\lambda$ a constant, and $H_{tot}=H_{harm}+H_{pert}$.

*Solution:*

a. Substituting $\Psi(x)$ into the S.E. yields
\begin{align*}
-\frac{\hbar^2}{2m}(4k^4x^2 - 2k^2) + \frac{1}{2}m\omega^2x^2 = E_0 \newline
\Rightarrow \left(\frac{1}{2}m\omega^2 - \frac{2k^4\hbar^2}{m}\right)x^2 = E_0 - \frac{\hbar^2 k^2}{m}
\end{align*}
which yields $k = \left(\frac{m^2\omega^2}{4\hbar^2}\right)^{1/4}$, as well as $E_0 = \frac{\hbar^2 k^2}{m} = \frac{\hbar\omega}{2}$.

b. We need $C$ to calculate $\Delta E$ exactly, which is set by normalization of $\Psi$.  I.e.
\begin{align*}
C^2 \int dx\;e^{-2k^2 x^2} = C^2 \sqrt{\frac{\pi}{2k^2}} = C^2\sqrt{\frac{\pi\hbar}{m\omega}} = 1
\end{align*}
so $\Psi(x) = \left(\frac{m\omega}{\pi\hbar}\right)^{1/4} e^{-(m\omega/2\pi\hbar)x^2}$.

The first-order correction to the energy is given by
\begin{align*}
\Delta E &= \langle\Psi\rvert H_{pert} \lvert\Psi\rangle \newline
&= 2\lambda \sqrt{\frac{m\omega}{\pi\hbar}} \int_0^{\infty} dx\;x e^{-(m\omega/\pi\hbar)x^2} \newline
&= \lambda \sqrt{\frac{m\omega}{\pi\hbar}} \left(\frac{\pi\hbar}{m\omega}\right) \newline
&= \lambda \sqrt{\frac{\pi\hbar}{m\omega}}
\end{align*}

**FA11, #6 (Perturbation theory)**

*Problem:* A hydrogen atom is placed in a time-dependent electric field $\vec{E} = E(t)\hat{k}$.  Consider transitions between the $n=1$ ground state and the $n=2$ excited states given the perturbation $H^{\prime} = eEz$.

a. Use symmetry to show that for all $n=1$ and $n=2$ states the matrix element $H^{\prime}_{ii}=0$.

b. Use symmetry arguments to show that only one of the $n=2$ states can be reached from the $n=1$ ground state.

c. Calculate the one remaining transition matrix element $H^{\prime}_{ij}$ between $n=1$ and $n=2$ states that is not zero.

*Solution:* The time-dependence of $\vec{E}$ is actually irrelevant here.  Since $\vec{E}$ always points in the $z$ direction, the associated perturbation is always odd in $z$, which is enough for the following symmetry arguments.

a. The diagonal elements $H^{\prime}_{ii}$ are $\langle i \rvert H^{\prime} \lvert i \rangle$.  Since $H^{\prime}$ is odd in $z$, these inner products must all be zero by symmetry (think of the integral representation, which always containts $\lvert\psi_n\rvert^2$.)

b. We need to look for a product of two states $\psi_m^* \psi_n$ that is odd in $z$, so that overall the integrand of the inner product is even.  The ground state is even in $z$, so we need an excited state that is odd in $z$.  Only the $n=2$, $l=1$, $m=0$ state is odd in $z$ since it is proportional to $\cos\theta$.

c. The one element which is non-zero:
\begin{align*}
H^{\prime}_{100,210} &= -eE\frac{1}{\pi a^3}\frac{1}{32\pi a^3}\frac{1}{a} \int d^3r\;e^{-r/a}e^{-2r/a} z^2 \newline
&= \ldots = -\left(\frac{2^8}{3^5\sqrt{2}}\right)eEa
\end{align*}

**FA09, #7 (Perturbation theory)**

*Problem:* Consider an electron in a hydrogen atom in a state with quantum numbers $n,l,m$.  Ignore electron spin for this problem.  Assume a uniform magnetic field $\vec{B}$ in the $z$ direction is slowly turned on.

a. Show that a possible magnetic vector potential is $\vec{A} = (\vec{B}\times\vec{r})/2$. *Addendum:* In which gauge are we working?

b. Show that the wavefunction for this electron in the presence of the magnetic field is the same as in the absence of the magnetic field, to lowest order in $\vec{A}$.  Use that $\vec{p} \rightarrow \vec{p} - (e/c)\vec{A}$.

c. Estimate the order of magnitude of $B$ for which the change in the wavefunction is no longer negligible.

d. Assuming there are $n_a$ atoms per unit volume in the same state, find an expression for the magnetic susceptibility per unit volume $\chi$ of this system.  Ignore the difference between $B$ and $H$.

Hint: $\chi = -\partial^2 E/\partial B^2$, with $E$ the energy per unit volume.

*Solution:*

a. Using some vector calculus identities,
\begin{align*}
\nabla \times A &= \frac{1}{2} \nabla \times (\vec{B}\times\vec{r}) \newline
&= \frac{1}{2}\left[\vec{B}(\nabla\cdot\vec{r}) - (\vec{B}\cdot\nabla)\vec{r}\right] \newline
&= \frac{1}{2}\left[\hat{z}B(3) - (B\partial_z)\vec{r}\right] \newline
&= B\hat{z}
\end{align*}

Note that $\nabla\cdot\vec{A}=0$; this is the Coulomb gauge.

b. The kinetic term in the Hamiltonian is affected by the addition of $\vec{B}$, namely
$$
\frac{p^2}{2m_e} \rightarrow \frac{(\vec{p}-\frac{e}{c}\vec{A})^2}{2m_e} = \frac{\vec{p}^2 + (e/c)^2\vec{A}^2 - (e/c)\vec{A}\cdot\vec{p}}{2m_e}
$$
(since $\vec{p}\cdot\vec{A} \sim \nabla\cdot\vec{A} = 0$ in this gauge).

Permute the product in the term of order $A$:
$$
\vec{A}\cdot\vec{p} = \frac{1}{2}(\vec{B}\times\vec{r})\cdot\vec{p} = \frac{1}{2}(\vec{r}\times\vec{p})\cdot{B} = \frac{1}{2}(\vec{L}\cdot\vec{B}) = \frac{1}{2}L_z B
$$
and since the original state was an eigenfunction of $L_z$, the wavefunction is unaffected by the addition of the magnetic field, to lowest order in $A$.

c. The $\vec{A}^2$ term is "important" when its ratio with the order $\vec{A}$ term is of order 1, roughly speaking. So
$$ \frac{e^2 A^2}{2m_e c^2} \frac{m_e c}{e A p} = \frac{eA}{2pc} \sim \frac{eBr}{pc} \sim \frac{eBr^2}{\hbar c} $$
Inserting the approximate size of the atom in the ground state ($r \sim 10^{-10} m$) yields $eA/2pc \sim 10^{-5}B\text{/T}$ (that's /Tesla units).  So only when $B \sim 10^5\text{T}$ does the order $\vec{A}^2$ contribution become significant.  (This is actually a very strong magnetic field).

d. Only the perturbation contains $B$, and hence only it contributes to $\chi$.  Additionally, only the order $\vec{A}^2$ will contribute since only it will survive $\partial_B^2$.  In the definition of $\chi$, $E$ is the average energy of the system, so for $n_a$ particles in the same state $\lvert\Psi\rangle$,
\begin{align*}
\chi &= -\partial_B^2 n_a \langle\Psi\rvert[(erB)^2/4m_e c^2]\lvert\Psi\rangle = -\frac{n_a e^2}{2 m_e c^2} \langle\Psi\rvert r^2 \lvert\Psi\rangle
\end{align*}

**FA09, #8 (Variational principle)**

*Problem:* Use the variational method with a Gaussian trial function $\psi(x) = e^{-ax^2/2}$ to estimate the ground state energy of a particle in the potential $V(x)=\lambda x^6$.

Hint: the integral $\int dx\;x^n e^{-ax^2} = \Gamma\left(\frac{n+1}{2}\right) a^{-(n+1)/2}$.

*Solution:* Computing $\langle H\rangle$ places an upper bound on the variational estimate for $E_0$.  First,
$$
\partial_x^2\psi(x) = \partial_x(-axe^{-ax^2/2}) = -ae^{-ax^2/2} + (ax)^2e^{-ax^2/2}
$$
so
\begin{align*}
\langle T\rangle &= -\frac{\hbar^2}{2m}\int dx\;\psi(x)\partial_x^2\psi(x) = \ldots = \frac{\hbar^2}{4m}\sqrt{\pi a} \newline
\langle V\rangle &= \lambda \int dx\;x^6\psi^2(x) = \ldots = \frac{15}{8}\lambda \sqrt{\frac{\pi}{a^7}}
\end{align*}

Now minimize $\langle H\rangle$ w.r.t. $a$:
\begin{align*}
\frac{\partial\langle H\rangle}{\partial a}\biggr\rvert_{a = a_{min}} &= 0 \newline
\Rightarrow a_{min} &= \left(\frac{105m\lambda}{2\hbar^2}\right)^{1/4}
\end{align*}
i.e. with this choice of $a$, we place the tighest possible upper bound on $E_0$ that we can using the variational principle.

**FA10, #8 (QHO)**

*Problem:* Compute $\langle\psi_0\rvert x^4 \lvert\psi_0\rangle$ in the ground state of the one-dimensional harmonic oscillator Hamiltonian $H = \frac{p^2}{2m} + \frac{1}{2} m\omega^2 x^2$.

*Solution:* Use the creation and annihilation operators for the harmonic oscillator, namely $a = (2m\omega\hbar)^{-1/2} (m\omega x-ip)$ (annihilation operator) and its Hermitian conjugate $a^{\dagger}$.  Solve for $x = \frac{1}{2}\sqrt{\frac{2\hbar}{m\omega}}(a + a^{\dagger})$.  The expansion of $x^4$ in $a$ and $a^{\dagger}$ thus has 16 terms, but the only ones that survive the inner product are those which do not have an $a$ on the right or an $a^{\dagger}$ on the left (which would kill the ground state ket/bra, respectively).  There should also be an equal number of creation and annihilation operators in any surviving terms, since the $H$ eigenstates are orthogonal.

With these considerations, the inner product becomes
\begin{align*}
\langle\psi_0\rvert x^4 \lvert\psi_0\rangle = \frac{1}{4}\left(\frac{\hbar}{m\omega}\right)^2 \langle\psi_0\rvert (aaa^{\dagger}a^{\dagger} + aa^{\dagger}aa^{\dagger}) \lvert\psi_0\rangle = \frac{3}{4}\left(\frac{\hbar}{m\omega}\right)^2
\end{align*}

**FA08, #5 (Other - measurements)**

*Problem:* Consider a quantum mechanical system which only has two available states, i.e. the ket $\lvert\psi\rangle$ is a vector in a 2-dimensional, complex vector space.  This space has a complete, orthonormal basis of kets $\lvert e_i\rangle$, for $i=1,2$.  The Hamiltonian of this system is $H = E_1\lvert e_1\rangle\langle e_1\rvert + E_2\lvert e_2\rangle\langle e_2\rvert$, where $E_1 < E_2$.  There is another observable $B$, with $B = b\lvert e_1\rangle\langle e_2\rvert + b\lvert e_1\rangle\langle e_2\rvert$, where $b$ is a positive real number.

The following experiments are performed in sequence:

- At time $t=t_0$, $B$ is measured.
- At time $t_2$, the energy is measured.
- At time $t_3$, $B$ is measured.
- At time $t_4$, $B$ is measured again.

a. Suppose that the outcome of experiment 1 is the larger of the two possible values.  Write the state for general time $t\in(t_1,t_2)$.

b. Suppose that at time $t_2$ the larger possible energy is measured.  Write the state for general $t\in(t_2,t_3)$.

c. Suppose that at time $t_3$ the *larger* possible value of $B$ is measured.  For what values (list all of them) of $\Delta t\equiv t_4-t_3$ is there a 100\% probability that the *smaller* value of $B$ will be measured at time $t_4$?

*Solution:*

a. Acting with $B$ sends the energy eigenstates $\lvert e_1\rangle \rightarrow \lvert e_2\rangle$ and $\lvert e_2\rangle \rightarrow \lvert e_1\rangle$.  The eigenstates of $B$ fulfill $B\lvert\lambda_B\rangle = \lambda_B\lvert\lambda_B\rangle$; in terms of the energy eigenstates, they are $\frac{1}{\sqrt{2}}(\lvert e_1 \rangle + \lvert e_2 \rangle)$ and $\frac{1}{\sqrt{2}}(\lvert e_1 \rangle - \lvert e_2 \rangle)$ with eigenvalues $+b$ and $-b$, respectively (let's call the states $\lvert +b \rangle$ and $\lvert -b \rangle$).  Note that, since $B$ does not commute with $H$, the $B$ eigenstates ARE NOT stationary states.

Thus, if at $t_1$ we measure $+b$, then in the time interval $t_1 < t < t_2$, the state will evolve as $\lvert\psi(t)\rangle = \frac{1}{\sqrt{2}}(e^{-iE_1t/\hbar}\lvert e_1\rangle + e^{-iE_2t/\hbar}\lvert e_2\rangle)$.

b. $\lvert e_1\rangle$ and $\lvert e_2\rangle$ are the eigenstates of $H$, so if the larger energy is measured, the state will remain $\lvert e_2\rangle$ until $t_3$ (up to an overall, undetectable phase).

c. Measuring $+b$ at $t=t_3$ again projects the system into the $B$ eigenstate $\lvert +b\rangle$.  The state at $t=t_4$ is
\begin{align*}
\lvert\psi(t_4)\rangle &= \frac{1}{\sqrt{2}} (e^{-iE_1\Delta t/\hbar}\lvert e_1\rangle + e^{-iE_2\Delta t/\hbar}\lvert e_2\rangle) 
\newline
&= \frac{1}{\sqrt{2}} e^{-iE_1\Delta t/\hbar}(\lvert e_1\rangle + e^{-i(E_2-E_1)\Delta t/\hbar}\lvert e_2\rangle)
\end{align*}
in which case the system will be in the eigenstate $\lvert -b\rangle$ every time $\Delta t = \frac{(2n+1)\pi\hbar}{E_2-E_1}$ for arbitrary $n\in\mathbb{Z}$.
