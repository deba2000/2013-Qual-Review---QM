### SP13, #5 (Spin)

***Problem:*** Consider a system of two spins that are coupled via a spin-spin interaction $W=a\mathbf{S}_1\cdot\mathbf{S}_2$.  Here $a$ is a measure of the coupling strength.  How many (spin) states does the two-particle system have, and what are their energy levels?  Show how you calculate this.

***Solution:*** In the $z$ spin basis, there are four possible states, each labeled by their total spin $S$ and spin $z$-projection $m_z$ as $\lvert S,m_z \rangle$.  (These are the two-particle states, which are product states of the individual spin states $\lvert \hbar/2,\hbar/2 \rangle = \lvert \uparrow \rangle$ and $\lvert \hbar/2,-\hbar/2 \rangle = \lvert \downarrow \rangle$).  They are the eigenstates of the mutually-commuting operators $\mathbf{S}^2$, $\mathbf{S}_1^2$, and $\mathbf{S}_2^2$ (where $\mathbf{S} = \mathbf{S}_1 + \mathbf{S}_2$).  The spin operators can be written in terms of the Paul spin matrices:
$$
\mathbf{S}_i = S_x^{(i)} + S_y^{(i)} + S_z^{(i)} = \frac{\hbar}{2} \left(\sigma_x^{(i)} + \sigma_y^{(i)} + \sigma_z^{(i)}\right).
$$
The following identities are useful for $z$ spinors:
$$
\begin{align}
\sigma_x \lvert \uparrow \rangle = \lvert \downarrow \rangle, \quad \sigma_x \lvert \downarrow \rangle = \lvert \uparrow \rangle, \\
\sigma_y \lvert \uparrow \rangle = -i \lvert \downarrow \rangle, \quad \sigma_y \lvert \downarrow \rangle = i\lvert \uparrow \rangle, \\
\sigma_z \lvert \uparrow \rangle = \lvert \uparrow \rangle, \quad \sigma_z \lvert \downarrow \rangle = \lvert \downarrow \rangle
\end{align}
$$

Now, the eigenstates:

**Triplet states** ($S=1$):
$$
\begin{align}
\lvert1,1\rangle &= \lvert\uparrow\uparrow\rangle \\
\lvert1,0\rangle &= \frac{1}{\sqrt{2}} \left(\lvert\uparrow\downarrow\rangle + \lvert\downarrow\uparrow\rangle\right) \\
\lvert1,-1\rangle &= \lvert\downarrow\downarrow\rangle
\end{align}
$$

**Singlet state** ($S=0$):
$$\lvert0,0\rangle = \frac{1}{\sqrt{2}}\left(\lvert\uparrow\downarrow\rangle - \lvert\downarrow\uparrow\rangle\right) $$

The interaction Hamiltonian is $W = a \mathbf{S}_1 \cdot \mathbf{S}_2 = a\left(S_x^{(1)}S_x^{(2)} + S_y^{(1)}S_y^{(2)} + S_z^{(1)}S_z^{(2)}\right)$. The Schrödinger equation for the first triplet state thus reads:
$$
\begin{align}
W\lvert\uparrow\uparrow\rangle &= a\left(S_x^{(1)}S_x^{(2)} + S_y^{(1)}S_y^{(2)} + S_z^{(1)}S_z^{(2)}\right) \lvert\uparrow\uparrow\rangle \\
&= a\frac{\hbar^2}{4} \left(\lvert\downarrow\downarrow\rangle + (i)(i)\lvert\downarrow\downarrow\rangle + \lvert\uparrow\uparrow\rangle\right) \\
&= a\frac{\hbar^2}{4} \lvert\uparrow\uparrow\rangle
\end{align}
$$
The result of $W\lvert\Psi\rangle$ for the other triplet states is the same, so all of the $S=1$ states have an interaction energy $E_{trip}=a\hbar^2/4$.

Now apply $W$ to the singlet state:
$$
\begin{align}
\frac{1}{\sqrt{2}} W \left(\lvert\uparrow\downarrow\rangle - \lvert\downarrow\uparrow\rangle\right) &= \frac{a}{\sqrt{2}} \left(S_x^{(1)}S_x^{(2)} + S_y^{(1)}S_y^{(2)} + S_z^{(1)}S_z^{(2)}\right) \left(\lvert\uparrow\downarrow\rangle - \lvert\downarrow\uparrow\rangle\right) \\
&= \frac{a\hbar^2}{4\sqrt{2}} \left( \lvert\downarrow\uparrow\rangle - \lvert\uparrow\downarrow\rangle + (i)(-i)\lvert\downarrow\uparrow\rangle - (-i)(i)\lvert\uparrow\downarrow\rangle - \lvert\uparrow\downarrow\rangle + \lvert\downarrow\uparrow\rangle \right) \\
&= \frac{-3a\hbar^2}{4\sqrt{2}} \left(\lvert\uparrow\downarrow\rangle - \lvert\downarrow\uparrow\rangle\right)
\end{align}
$$

so the interaction energy of the singlet state is $-3a\hbar^2/4$.

### SP12, #6 (Reflection/transmission)

***Problem:*** A one-dimensional non-relativistic particle interacts with the potential $V=\lambda\frac{\hbar^2}{2m}\delta(x)$.

1. Calculate the reflection and transmission coefficients (probabilities) as a function of the incident particle wavenumber $k$.
2. Calculate the scattering and bound states for $\lambda < 0$.  Show that there is a single bound state, and that it is orthogonal to the scattering states.

***Solution***:
The (time-independent) Schrödinger equation with this potential is solved in the regions $x\neq0$ by the plane-wave states
$$
\begin{align}
\psi(x) = \begin{cases}
A e^{ikx} + B e^{-ikx} & \text{if } \; x < 0 \\
C e^{ikx} + D e^{-ikx} & \text{if } \; x > 0
\end{cases}
\end{align}
$$

One boundary condition is that the wavefunction be continuous at $x=0$ (as it should be everywhere); this enforces $\boxed{A + B = C + D}$.  The second boundary condition we use is the discontinuity in $\psi^{\prime}(x=0)$ (remember, $\partial_x\psi$ is normally continuous, *except* where the potential goes to infinity).  The discontinuity is found by integration (by parts) of the S.E. around $x=0$:
$$
\begin{align}
-\frac{\hbar^2}{2m} \int_{x=0^-}^{x=0^+} dx \: \partial^2_x \psi(x) &= \lambda \frac{\hbar^2}{2m} \int_{x=0^-}^{x=0^+} dx \: \psi(x) \delta(x) \\
\partial_x\psi(x=0^+) - \partial_x\psi(x=0^-) &= -\lambda\psi(x=0).
\end{align}
$$
Substituting in our solution gives the second boundary condition, $\boxed{-\lambda(A+B) = ik(C-D-A+B)}$.

The scattering states (for which $E>0$) satisfy $\partial_x^2 \psi(x) = -k^2 \psi(x)$ (where $k \equiv \sqrt{2mE}/\hbar$) for $x \neq 0$.  Unforunately, the plane wave states are not normalizable, so there will always be one constant we can't eliminate.  However, we can at least get rid of $D$ if we imagine sending in particles from $x=-\infty$, and none from the right.  With this, BC 1 & 2 are solved for $A$, leaving
$$
\begin{align}
B &= \frac{i\lambda}{2k-i\lambda}A \\
C &= \left[1 + \frac{i\lambda}{2k-i\lambda}\right]A
\end{align}
$$

The best we can do, then, is write
$$
\psi(x) = \begin{cases}
A\left(e^{ikx} + \frac{i\lambda}{2k-i\lambda}e^{-ikx}\right), & \text{if }\;x<0 \\
A\left(1 + \frac{i\lambda}{2k-i\lambda}\right)e^{ikx}, &\text{if }\;x>0
\end{cases}
$$

After some algebra, the reflection and transmission coefficients are:
$$
\begin{align}
R(k) &\equiv \frac{\lvert B \rvert^2}{\lvert A \rvert^2} = \frac{\lambda^2}{(2k)^2+\lambda^2} = \frac{1}{(2k/\lambda)^2 + 1} \\
T(k) &\equiv \frac{\lvert C \rvert^2}{\lvert A \rvert^2} = \frac{(2k)^2}{(2k)^2+\lambda^2} = \frac{1}{1 + (\lambda/2k)^2}
\end{align}
$$
Note the limits: when $2k/\lambda \ll 1$, $R\rightarrow 1$ and $T\rightarrow 0$ (in the low-momentum limit, all particles are reflected).  The opposite effect occurs in the high-momentum limit, when $2k/\lambda \gg 1$.

The bound state satisfies $E<0$.  In the region $x<0$, the S.E. reads:
$$ \partial_x^2 \psi(x) = \kappa^2 \psi(x); \quad \kappa \equiv \frac{\sqrt{-2mE}}{\hbar} $$
(renamed $k \rightarrow \kappa$ to avoid confusion with the scattering states later).  The general solution to this equation is
$$
\begin{align}
\psi(x) = \begin{cases}
A e^{\kappa x} + B e^{-\kappa x} & \text{if } \; x < 0 \\
C e^{\kappa x} + D e^{-\kappa x} & \text{if } \; x > 0
\end{cases}
\end{align}
$$
where we're forced to set $B=0$ and $C=0$, so that $\psi$ doesn't blow up at $x=\pm\infty$.  Using the continuity of $\psi$ at $x=0$ as the first boundary condition sets $A=D$:
$$
\psi(x) = \begin{cases}
Ae^{\kappa x}, & x<0 \\
Ae^{-\kappa x}, & x>0
\end{cases}
$$

Again using the discontinuity in $\partial_x\psi$ at $x=0$ as the second B.C. gives $\kappa = \lambda/2$.  Finally, to eliminate $A$, normalize the wavefunction to find the following bound state:
$$
\psi(x) = \begin{cases}
\sqrt{-\lambda/2}e^{-\lambda x/2}, & x<0 \\
\sqrt{-\lambda/2}e^{\lambda x/2}, & x>0
\end{cases}
$$
(note the choice of signs since $\lambda<0$).

### SP11, #5 (Entanglement)

***Problem:*** Consider two entangled spin $1/2$ particles separated by a large distance.  The four Bell states are given as
$$
\begin{align}
\lvert\Phi_{\pm}^{(AB)}\rangle &= \frac{1}{\sqrt{2}} \left[\lvert z+,z+\rangle \pm \lvert z-,z-\rangle\right] \\
\lvert\Psi_{\pm}^{(AB)}\rangle &= \frac{1}{\sqrt{2}} \left[\lvert z+,z-\rangle \pm \lvert z-,z+\rangle\right]
\end{align}
$$
where $\lvert z\pm\rangle$ are the two states along or against the $z$ axis; the first slot refers to particle $A$ which is nearby, the second to particle $B$ which is far away.  Consider applying a magnetic field only at location $A$, either in the $\hat{x}$ or $\hat{z}$ directions, for a time $t$.

a. Write down the Hamiltonians, $H_z$ and $H_x$, for the two operations above, and also write down the time evolution operators $U_z(t,0)$ and $U_x(t,0)$.

b. Find a time duration $t_A$ which results in $U_z\lvert z+\rangle = -i\lvert z+\rangle$.

c. Using time $t_A$, find $U_z\lvert z-\rangle$ and $U_x\lvert z\pm\rangle$.  (Hint: you may need to change from the $z$ basis to the $x$ basis).

d. Using the two operations above, show that any one of the Bell states can be transformed into any other, when operating only at location $A$.

***Solution:***

a. The Hamiltonian for a spin in an external field is $H=-\vec{\mu}\cdot\vec{B}$.  Since only particle $A$ feels the external field, $H_x = -\gamma B_x S_x$ (with $\vec{B}$ applied in the $x$ direction) and $H_z = -\gamma B_z S_z$, up to some proportionality constant $\alpha$.  (Actually $\gamma = q/mc$ is the gyromagnetic ratio).

The time evolution operator is $U(t,0)=e^{-iHt/\hbar}$; therefore $U_x(t,0) = e^{i\gamma B_x S_x t/\hbar}$ and $U_z(t,0) = e^{i\gamma B_z S_z t/\hbar}$.

b. Apply the time evolution operator to the state:
$$
\begin{align}
U_z(t,0)\lvert z+\rangle &= e^{i\gamma B_z S_z t/\hbar} \lvert z+\rangle \\
&= \left(1 + \frac{i\gamma B_z S_z t}{\hbar} + \ldots\right) \lvert z+\rangle \\
&= \left(1 + \frac{i\gamma B_z t}{2} + \ldots\right) \lvert z+\rangle \\
&= e^{i\gamma B_z t/2}\lvert z+\rangle.
\end{align}
$$
We want $e^{i\gamma B_z t_A/2}=-i$, so $t_A = \frac{(4n+3)\pi}{\gamma B_z}$ for arbitrary integer $n\geq 0$.

c. $U_z(t_A,0)\lvert z-\rangle = e^{i\gamma B_z S_z t_A/\hbar} \lvert z-\rangle = e^{-i\gamma B_z t_A/2}\lvert z-\rangle = e^{-i(4n+3)\pi/2}\lvert z-\rangle = i\lvert z-\rangle$.

Rewrite $\lvert z\pm\rangle = \frac{1}{\sqrt{2}} \left[\lvert x+\rangle \pm \lvert x-\rangle\right]$ to find that $U_x(t_A,0)\lvert z\pm\rangle = -i\lvert z\mp\rangle$.

d. $U_z(t_A,0)\lvert\Phi_+^{(AB)}\rangle = -i\lvert\Phi_-^{(AB)}\rangle$, and so on.

### SP11, #6 (Perturbation theory)

***Problem:***

a. The ground state wave function of the one dimensional harmonic oscillator ($H_{harm}=\frac{p^2}{2m}+\frac{1}{2}m\omega^2 x^2$) has the Gaussian form $\Psi(x)=Ce^{-k^2 x^2}$.  Using the Schrodinger equation, find $k$, and also find the ground state energy $E_0$.  Show all steps.

b. Use first order perturbation theory to find the first order correction $\Delta E$ to the ground state energy given by a perturbation to the potential, $H_{pert}=\lambda \lvert x \rvert$, with $\lambda$ a constant, and $H_{tot}=H_{harm}+H_{pert}$.

***Solution:***

a. Substituting $\Psi(x)$ into the S.E. yields
$$
\begin{align}
-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2}\left(C e^{-k^2 x^2}\right) = E_0 C e^{-k^2 x^2} \\
-\frac{\hbar^2}{2m}(4k^4x^2 - 2k^2) + \frac{1}{2}m\omega^2x^2 = E_0 \\
\Rightarrow \left(\frac{1}{2}m\omega^2 - \frac{2k^4\hbar^2}{m}\right)x^2 = E_0 - \frac{\hbar^2 k^2}{m}.
\end{align}
$$
The term in parenthesis on the LHS must independently go to zero, so that this equation is true at *all* $x$.  This yields $k = \left(\frac{m^2\omega^2}{4\hbar^2}\right)^{1/4}$.  The RHS yields $E_0 = \frac{\hbar^2 k^2}{m} = \frac{\hbar\omega}{2}$.

b. We need $C$ to calculate $\Delta E$ exactly, which is set by normalization of $\Psi$:
$$
C^2 \int_{-\infty}^\infty dx\;e^{-2k^2 x^2} = C^2 \sqrt{\frac{\pi}{2k^2}} = C^2\sqrt{\frac{\pi\hbar}{m\omega}} = 1 \quad \Rightarrow \quad C = \left(\frac{m\omega}{\pi\hbar}\right)^{1/4}
$$
so $\Psi(x) = \left(\frac{m\omega}{\pi\hbar}\right)^{1/4} e^{-(m\omega/2\pi\hbar)x^2}$.

The first-order correction to the energy is given by
$$
\begin{align}
\Delta E &= \langle\Psi\rvert H_{pert} \lvert\Psi\rangle \\
&= \lambda \sqrt{\frac{m\omega}{\pi\hbar}} \int_{-\infty}^{\infty} dx\; \lvert x\rvert e^{-(m\omega/\pi\hbar)x^2} \\
&= 2\lambda \sqrt{\frac{m\omega}{\pi\hbar}} \int_0^{\infty} dx\; x e^{-(m\omega/\pi\hbar)x^2} \\
&= 2\lambda \sqrt{\frac{m\omega}{\pi\hbar}} \int_0^{\infty} dx\; \left(-\frac{\pi\hbar}{2m\omega}\right) \frac{\partial}{\partial x} e^{-(m\omega/\pi\hbar)x^2} \\
&= \lambda \sqrt{\frac{m\omega}{\pi\hbar}} \left(-\frac{\pi\hbar}{m\omega}\right) e^{-(m\omega/\pi\hbar)x^2}\rvert_0^\infty \\
&= \lambda \sqrt{\frac{m\omega}{\pi\hbar}} \left(\frac{\pi\hbar}{m\omega}\right) = \lambda \sqrt{\frac{\pi\hbar}{m\omega}}.
\end{align}
$$

### FA11, #6 (Perturbation theory)

***Problem:*** A hydrogen atom is placed in a time-dependent electric field $\vec{E} = E(t)\hat{k}$.  Consider transitions between the $n=1$ ground state and the $n=2$ excited states given the perturbation $H^{\prime} = eEz$.

a. Use symmetry to show that for all $n=1$ and $n=2$ states the matrix element $H^{\prime}_{ii}=0$.

b. Use symmetry arguments to show that only one of the $n=2$ states can be reached from the $n=1$ ground state.

c. Calculate the one remaining transition matrix element $H^{\prime}_{ij}$ between $n=1$ and $n=2$ states that is not zero.

***Solution:*** The time-dependence of $\vec{E}$ is actually irrelevant here.  Since $\vec{E}$ always points in the $z$ direction, the associated perturbation (potential) is always odd in $z$, which is enough for the following symmetry arguments.

a. The diagonal elements $H^{\prime}_{ii}$ are $\langle i \rvert H^{\prime} \lvert i \rangle$.  Since $H^{\prime}$ is odd in $z$, these inner products must all be zero by symmetry.  One way to see this is to imagine the integral representation of the matrix element: it will contain $\lvert\psi\rvert^2$, which is always even.  If there is no integral representation, we can resort to the using the Wigner-Eckhardt theorem...

b. We need to look for a product of two states $\psi_m^* \psi_n$ that is odd in $z$, so that overall the integrand of the inner product is even.  The ground state is even in $z$, so we need to look for an excited state that is odd in $z$.

Think of the spherical harmonic representations of the states.  The $n=2$, $l=0$ state is even (no angular momentum $\rightarrow$ spherically symmetric state), so that's out.  The $n=2$, $l=1$, $m=0$ state is odd in $z$, because it's proportional to $\cos\theta$.  The $n=2$, $l=1$, $m=\pm 1$ states, on the other hand, go like $\sin\theta$ and are thus even in $z$.  So of all the first excited states, *only* the $n=2$, $l=1$, $m=0$ state is reachable under this perturbation.

c. The one element which is non-zero (using $z = r\cos\theta$, and some parametric differentiation)
$$
\begin{align}
H^{\prime}_{100,210} &= -eE\frac{1}{\sqrt{\pi a^3}}\frac{1}{\sqrt{32\pi a^5}} \int d^3r\; \left(e^{-r/a}\right) \left(r e^{-r/2a}\cos\theta\right) z \\
&= -eE\frac{2\pi}{\sqrt{32 \pi^2 a^8}} \int_0^\pi d\theta \: \sin\theta \cos^2\theta \int_0^\infty dr \: r^4 e^{-3r/2a} \\
&= -eE\frac{1}{\sqrt{8}a^4} \left[-\frac{1}{3}\cos^3\theta\right]_0^\pi \frac{\partial^4}{\partial \beta^4} \int_0^\infty dr \; e^{-r\beta} \\
&= -\frac{2eE}{3\sqrt{8}a^4} \frac{\partial^4}{\partial \beta^4} \frac{1}{\beta} \\
&= -\frac{eE}{3\sqrt{2}a^4} \frac{4!}{\beta^5} \\
&= -\frac{eE}{3\sqrt{2}a^4} \frac{(4!)(2a)^5}{3^5} = -\frac{256}{243\sqrt{2}} eEa.
\end{align}
$$
I'm probably off by some constant factor...

### FA09, #7 (Perturbation theory)

***Problem:*** Consider an electron in a hydrogen atom in a state with quantum numbers $n,l,m$.  Ignore electron spin for this problem.  Assume a uniform magnetic field $\vec{B}$ in the $z$ direction is slowly turned on.

a. Show that a possible magnetic vector potential is $\vec{A} = (\vec{B}\times\vec{r})/2$.

b. Show that the wavefunction for this electron in the presence of the magnetic field is the same as in the absence of the magnetic field, to lowest order in $\vec{A}$.  Use that $\vec{p} \rightarrow \vec{p} - (e/c)\vec{A}$.

c. Estimate the order of magnitude of $B$ for which the change in the wavefunction is no longer negligible.

d. Assuming there are $n_a$ atoms per unit volume in the same state, find an expression for the magnetic susceptibility per unit volume $\chi$ of this system.  Ignore the difference between $B$ and $H$.

Hint: $\chi = -\partial^2 E/\partial B^2$, with $E$ the energy per unit volume.

***Solution:***

a. Using the vector triple-product identity,
$$
\begin{align}
\nabla \times \vec{A} &= \frac{1}{2} \nabla \times (\vec{B}\times\vec{r}) \\
&= \frac{1}{2}\left[(\vec{r}\cdot\nabla)\vec{B} - (\vec{B}\cdot\nabla)\vec{r} + \vec{B}(\nabla\cdot\vec{r}) - \vec{r}(\nabla\cdot\vec{B}) \right] \\
\end{align}
$$
The field is uniform, so the first term is 0.  Additionally, $\vec{B}$ is divergenceless so the fourth term is zero.  Thus,
$$
\begin{align}
\nabla \times \vec{A} &= \frac{1}{2}\left[-(B\partial_z)\vec{r} + \hat{z}B(3)\right] \\
&= B\hat{z}
\end{align}
$$

b. Only the kinetic term in the Hamiltonian is affected by the addition of $\vec{B}$ by entering into the canonical momentum of the particle:
$$
\frac{p^2}{2m_e} \rightarrow \frac{(\vec{p}-\frac{e}{c}\vec{A})^2}{2m_e} = \frac{\vec{p}^2 + (e/c)^2\vec{A}^2 - (e/c) \left(\vec{p}\cdot\vec{A} + \vec{A}\cdot\vec{p}\right)}{2m_e}.
$$
However, $\vec{p}\cdot\vec{A} \sim \nabla\cdot\vec{A} = 0$ in this gauge, so the second-to-last term is gone.

Permute the product in the term of order $A$:
$$
\vec{A}\cdot\vec{p} = \frac{1}{2}(\vec{B}\times\vec{r})\cdot\vec{p} = \frac{1}{2}(\vec{r}\times\vec{p})\cdot{B} = \frac{1}{2}(\vec{L}\cdot\vec{B}) = \frac{1}{2}L_z B
$$
and since the original state was an eigenfunction of $L_z$, the wavefunction is unaffected by the addition of the magnetic field, to lowest order in $A$ (although the energy eigenvalue is certainly changed).

c. The $\vec{A}^2$ term is "important" when its ratio with the order $\vec{A}$ term is of order 1, roughly speaking. So
$$ \frac{e^2 A^2}{2m_e c^2} \frac{2m_e c}{e A p} = \frac{eA}{pc} \sim \frac{eBr}{pc} \sim \frac{eBr^2}{\hbar c} $$
Inserting the approximate size of the atom in the ground state ($r \sim 10^{-10}\text{ m}$) yields $eBr^2/\hbar c \sim B (10^{-5}\text{ T}^{-1})$.  So only when $B \sim 10^5\text{ T}$ does the order $\vec{A}^2$ contribution become significant.  (This is actually a very strong magnetic field).

d. Only the perturbation contains $B$, and hence only it contributes to $\chi$.  Additionally, only the order $\vec{A}^2$ will contribute to $\chi$ because of the second derivative with respect to $B$.  In the definition of $\chi$, $E$ is the average energy of the system, so for $n_a$ particles in the same state $\lvert\Psi\rangle$,
$$
\chi = -n_a \frac{\partial^2}{\partial B^2} \langle\Psi\rvert[(erB)^2/4m_e c^2]\lvert\Psi\rangle = -\frac{n_a e^2}{2 m_e c^2} \langle\Psi\rvert r^2 \lvert\Psi\rangle = -\frac{3 n_a a_0^2}{2 m_e c^2}
$$
if, for example, all the atoms are in the ground state.

### FA09, #8 (Variational principle)

***Problem:*** Use the variational method with a Gaussian trial function $\psi(x) = e^{-ax^2/2}$ to estimate the ground state energy of a particle in the potential $V(x)=\lambda x^6$.

Hint: the integral $\int dx\;x^n e^{-ax^2} = \Gamma\left(\frac{n+1}{2}\right) a^{-(n+1)/2}$.

***Solution:*** Computing $\langle H\rangle$ places an upper bound on the variational estimate for $E_0$.  First,
$$
\partial_x^2\psi(x) = \partial_x(-axe^{-ax^2/2}) = -ae^{-ax^2/2} + (ax)^2e^{-ax^2/2}
$$
so
$$
\begin{align}
\langle T\rangle &= -\frac{\hbar^2}{2m}\int dx\;\psi(x)\partial_x^2\psi(x) = \ldots = \frac{\hbar^2}{4m}\sqrt{\pi a} \\
\langle V\rangle &= \lambda \int dx\;x^6\psi^2(x) = \ldots = \frac{15}{8}\lambda \sqrt{\frac{\pi}{a^7}}
\end{align}
$$

Now minimize $\langle H\rangle$ w.r.t. $a$:
$$
\begin{align}
\frac{\partial\langle H\rangle}{\partial a}\biggr\rvert_{a = a_{min}} &= \frac{\partial(\langle T\rangle + \langle V\rangle)}{\partial a}\biggr\rvert_{a = a_{min}} = 0 \\
&= \frac{\hbar^2}{8m} \sqrt{\frac{\pi}{a_{min}}} - \frac{105}{16}\lambda\sqrt{\frac{\pi}{a_{min}^9}} \\
\Rightarrow \quad  &a_{min} = \left(\frac{105m\lambda}{2\hbar^2}\right)^{1/4}
\end{align}
$$
This choice of $a$ gives the lowest possible upper bound on $E_0$:
$$
E_0 \leq \frac{\hbar^2}{4} \sqrt{\pi a_{min}} + \frac{15}{8}\lambda \sqrt{\frac{\pi}{a_{min}^7}}.
$$

### FA10, #8 (QHO)

***Problem:*** Compute $\langle\psi_0\rvert x^4 \lvert\psi_0\rangle$ in the ground state of the one-dimensional harmonic oscillator Hamiltonian $H = \frac{p^2}{2m} + \frac{1}{2} m\omega^2 x^2$.

***Solution:*** Use the creation and annihilation operators for the harmonic oscillator, namely $a = (2m\omega\hbar)^{-1/2} (m\omega x-ip)$ (annihilation operator) and its Hermitian conjugate $a^{\dagger}$.  Solve for $x = \frac{1}{2}\sqrt{\frac{2\hbar}{m\omega}}(a + a^{\dagger})$.

The expansion of $x^4$ in $a$ and $a^{\dagger}$ thus has 16 terms, each containing combinations of 4 creation or annihilation operators.  But, the only ones that survive the inner product are those which do not have an $a$ on the right or an $a^{\dagger}$ on the left (which would kill the ground state ket/bra, respectively).  There must also be an equal number of creation and annihilation operators in any surviving terms, since the $H$ eigenstates are orthogonal (otherwise the inner product would be zero).

With these considerations, the inner product becomes
$$
\begin{align}
\langle\psi_0\rvert x^4 \lvert\psi_0\rangle &= \frac{1}{16}\left(\frac{2\hbar}{m\omega}\right)^2 \langle\psi_0\rvert (aaa^{\dagger}a^{\dagger} + aa^{\dagger}aa^{\dagger}) \lvert\psi_0\rangle \\
&= \frac{1}{4}\left(\frac{\hbar}{m\omega}\right)^2 \langle\psi_0\rvert \left(\sqrt{1 \cdot 2 \cdot 2 \cdot 1} + \sqrt{1 \cdot 1 \cdot 1 \cdot 1}\right) \lvert\psi_0\rangle \\
&= \frac{3}{4}\left(\frac{\hbar}{m\omega}\right)^2
\end{align}
$$

### FA08, #5 (Other - measurements)

***Problem:*** Consider a quantum mechanical system which only has two available states, i.e. the ket $\lvert\psi\rangle$ is a vector in a 2-dimensional, complex vector space.  This space has a complete, orthonormal basis of kets $\lvert e_i\rangle$, for $i=1,2$.  The Hamiltonian of this system is $H = E_1\lvert e_1\rangle\langle e_1\rvert + E_2\lvert e_2\rangle\langle e_2\rvert$, where $E_1 < E_2$.  There is another observable $B$, with $B = b\lvert e_1\rangle\langle e_2\rvert + b\lvert e_2\rangle\langle e_1\rvert$, where $b$ is a positive real number.

The following experiments are performed in sequence:

- At time $t=t_0$, $B$ is measured.
- At time $t_2$, the energy is measured.
- At time $t_3$, $B$ is measured.
- At time $t_4$, $B$ is measured again.

a. Suppose that the outcome of experiment 1 is the larger of the two possible values.  Write the state for general time $t\in(t_1,t_2)$.

b. Suppose that at time $t_2$ the larger possible energy is measured.  Write the state for general $t\in(t_2,t_3)$.

c. Suppose that at time $t_3$ the *larger* possible value of $B$ is measured.  For what values (list all of them) of $\Delta t\equiv t_4-t_3$ is there a 100% probability that the *smaller* value of $B$ will be measured at time $t_4$?

***Solution:***

a. The eigenstates of $B$ fulfill $B\lvert\lambda_B\rangle = \lambda_B\lvert\lambda_B\rangle$, but we want these states in the energy eigenket basis.  Note that acting with $B$ sends the energy eigenstates $\lvert e_1\rangle \rightarrow \lvert e_2\rangle$ and $\lvert e_2\rangle \rightarrow \lvert e_1\rangle$.  Thus, the matrix representation of the eigenvalue equation for $B$ (in the energy eigenbasis) is
$$
\begin{pmatrix}
0 & b \\
b & 0
\end{pmatrix}
\begin{pmatrix}
c_1 \\ c_2
\end{pmatrix}
=
\lambda_B
\begin{pmatrix}
c_1 \\ c_2
\end{pmatrix}
$$
This matrix has eigenvalues $b$ and $-b$, with corresponding eigenvectors $\frac{1}{\sqrt{2}} (1, 1)$ and $\frac{1}{\sqrt{2}} (1,-1)$.  Thus, in terms of the energy eigenstates, the eigenstates of $B$ are $\frac{1}{\sqrt{2}}(\lvert e_1 \rangle + \lvert e_2 \rangle)$ and $\frac{1}{\sqrt{2}}(\lvert e_1 \rangle - \lvert e_2 \rangle)$ (let's call the states $\lvert +b \rangle$ and $\lvert -b \rangle$).

If we measure $B=+b$ at $t=t_0$, then $\lvert\psi(t=0)\rangle = \frac{1}{\sqrt{2}} \left(\lvert e_1\rangle + \lvert e_2\rangle\right)$.  Note that, since $[B,H] \neq 0$, the $B$ eigenstates ARE NOT stationary states.  Thus, over the time interval $t_1 < t < t_2$, the state will evolve as $\lvert\psi(t)\rangle = \frac{1}{\sqrt{2}}(e^{-iE_1t/\hbar}\lvert e_1\rangle + e^{-iE_2t/\hbar}\lvert e_2\rangle)$.

b. $\lvert e_1\rangle$ and $\lvert e_2\rangle$ are the eigenstates of $H$ (thus they are stationary states), so if the larger energy is measured, the state will be $\lvert e_2\rangle$ until $t_3$ (up to an overall, undetectable phase $e^{-iE_2(t-t_2)/\hbar}$).

c. Measuring $+b$ at $t=t_3$ again projects the system into the $B$ eigenstate $\lvert +b\rangle$.  The state at $t=t_4$ is
$$
\begin{align}
\lvert\psi(t_4)\rangle &= \frac{1}{\sqrt{2}} (e^{-iE_1\Delta t/\hbar}\lvert e_1\rangle + e^{-iE_2\Delta t/\hbar}\lvert e_2\rangle) \\
&= \frac{1}{\sqrt{2}} e^{-iE_1\Delta t/\hbar}(\lvert e_1\rangle + e^{-i(E_2-E_1)\Delta t/\hbar}\lvert e_2\rangle)
\end{align}
$$
in which case the system will be in the eigenstate $\lvert -b\rangle$ every time $\Delta t = \frac{(2n+1)\pi\hbar}{E_2-E_1}$ for arbitrary integer $n \geq 0$.
